```python
import pandas as pd
```

```python
df_movies = pd.read_csv("movies.csv")
df_financials = pd.read_csv("financials.csv")
df_languages = pd.read_csv("languages.csv")
```

```python
df_movies.head(3)
```

```python
df_financials.head(3)
```

```python
df_languages.head(3)
```

```python
df_new_movies = pd.read_csv("new_movies.csv")
df_new_movies
```

```python
df_movies = pd.concat([df_movies,df_new_movies],ignore_index=True)
df_movies.tail(5)
```

```python
df_movies = pd.merge(df_movies,df_languages,on="language_id")
df_movies
```

```python
df_movies.head(5)
```

```python
df_movies = pd.merge(df_movies,df_financials,on="movie_id",how='left')
df_movies
```

```python
df_movies.to_csv("final_complete_data.csv" , index=False ,columns = ['movie_id','title','lang_name','budget','revenue','currency'])
```
